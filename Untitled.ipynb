```julia
using Statistics
using QuantumCumulants
using ModelingToolkit, OrdinaryDiffEq
using Distributions
using Plots
```

```julia
function tensor_cavs(N)
    Hlist = [FockSpace(Symbol(:cavity,i)) for i=1:N]
    return ⊗(Hlist...)
end
N=1
h = tensor_cavs(N)
a(i) = Destroy(h, Symbol(:a, i), i)
a_dag(i) = Create(h, Symbol(:a, i), i)
@cnumbers Δ γ J U F
H = sum(F*(a_dag(i)+a(i)) -Δ*a_dag(i)*a(i) for i=1:N)
D = [a(i) for i=1:N]
rates = [γ for i=1:N]
H
```

```julia
#First order equations
eqs_1st = meanfield([a(i) for i in 1:N], H, D; rates=rates, order=1)
eqs_completed_1st = complete(eqs_1st)
```

```julia
#Numerical resolution of the mean-field equations
using ModelingToolkit, OrdinaryDiffEq
@named sys = ODESystem(eqs_completed_1st)
p0 = (Δ=>10, J=>0.0, γ=>0.01, U=>0, F=>5)

list = [2, -1, 1im, 2im,2+1im,-1-1im ,2-1im,1+1im]
list = [2]
l = length(list)

u0 = zeros(ComplexF64, length(eqs_completed_1st))

for i in 1:Int(N)
 #  num = mod(i, 4)
  #  if(num == 0)
  #      num=4
  #  end
    #u0[i] = list[num]
    u0[i] = list[i]
end
print(u0)
```

```julia
t_lim = 1.0
prob = ODEProblem(sys,u0,(0.0,t_lim),p0)
sol = solve(prob,Vern7(), dense=false, saveat = collect(range(0,t_lim,step=0.01))  )
print("ciao")
print(p0)
print(u0)
```

```julia
plot(sol.t, real.(sol[a(1)]), xlabel="t", ylabel="a", label="real cav 1")
plot!(sol.t, imag.(sol[a(1)]), xlabel="t", ylabel="a", label="imag cav 1")
```
